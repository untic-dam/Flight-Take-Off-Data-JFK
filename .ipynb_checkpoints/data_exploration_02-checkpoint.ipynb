#### One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder

X = df_obj 
X = X.dropna()
encoder = OneHotEncoder(handle_unknown='ignore')
X_ohe = encoder.fit_transform(X) #retourne une matrice simplifiée
```

```python
X_ohe
```

#### Dimension Reduction

```python
y = df['TAXI_OUT']
y.shape
```

```python
from sklearn.decomposition import PCA
from sklearn.decomposition import TruncatedSVD

n = X_ohe.shape[1] - 1

model = TruncatedSVD(n_components=n) # n_components = nombres de dimension
X_reduced = model.fit_transform(X_ohe)

cums = np.cumsum(model.explained_variance_ratio_)
plt.plot(cums)
```

```python
cums_99 = np.argmax(np.cumsum(model.explained_variance_ratio_)>0.95)
print(cums_99)
print('{:.0%}'.format(cums_99/n))
```

```python
model = TruncatedSVD(n_components=cums_99)
X_reduced = model.fit_transform(X_ohe)
```

```python
X_reduced.shape
```

```python
df_obj.head()
```

```python
from sklearn.preprocessing import LabelEncoder
y = df['OP_UNIQUE_CARRIER']
transformer = LabelEncoder()
y = transformer.fit_transform(y)
model = TruncatedSVD(n_components=3)
X_reduced = model.fit_transform(X_ohe)
plt.scatter(X_reduced[:,0], X_reduced[:,1], c=y)
plt.colorbar();
```

```python
from sklearn.preprocessing import LabelEncoder
y = df['Condition']
transformer = LabelEncoder()
y = transformer.fit_transform(y)
model = TruncatedSVD(n_components=3)
X_reduced = model.fit_transform(X_ohe)
plt.scatter(X_reduced[:,0], X_reduced[:,1], c=y)
plt.colorbar();
```

### Feature Selection

```python
from sklearn.feature_selection import RFE, RFECV
from sklearn.linear_model import LinearRegression

X = df_num.drop(labels=['TAXI_OUT'], axis=1)
y = df_num['TAXI_OUT']

X.shape
```

```python
selector = RFECV(LinearRegression(), step=2, min_features_to_select=7, cv=3)
selector.fit_transform(X,y)
```

```python
np.array(X.columns)[selector.get_support()]
```

### Variance

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=10)
selector.fit_transform(X)
selector.get_support() # colonne/features selectionnées
for i in np.array(X.columns)[selector.get_support()]:
    print(i)
```

```python
X.var(axis=0)
```

### transform num cols

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_std = scaler.fit_transform(X)
X_std.var(axis=0)
```

```python
X_std.shape
```

```python
X.columns
```

```python
selector = VarianceThreshold(threshold=0.05)
selector.fit_transform(X_std)
selector.get_support() # colonne/features selectionnées
for i in np.array(X.columns)[selector.get_support()]:
    print(i)
```
